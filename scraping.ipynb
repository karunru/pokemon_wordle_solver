```python
import requests
from bs4 import BeautifulSoup
import os
import time
import numpy as np
import pandas as pd
```

```python
proxies = {"http": os.environ["ASKUL_PROXY"], "https": os.environ["ASKUL_PROXY"]}
```

```python
i = 1
res = requests.get(f"https://zukan.pokemon.co.jp/detail/{i}", proxies=proxies)
res
```

```python
#  HTML  BeautifulSoup 
soup = BeautifulSoup(res.content, 'html.parser')
soup.find('title').text
```

```python
titles =  []
```

```python
while res.status_code == 200:
    res = requests.get(f"https://zukan.pokemon.co.jp/detail/{i}", proxies=proxies)
    soup = BeautifulSoup(res.content, "html.parser")
    title = soup.find("title").text
    titles.append(title)
    print(f"{i=}, {title=}")
    i = i+1
    time.sleep(np.random.randint(low=1, high= 999)/1000)
```

```python
i = 655
while res.status_code == 200:
    res = requests.get(f"https://zukan.pokemon.co.jp/detail/{i}", proxies=proxies)
    soup = BeautifulSoup(res.content, "html.parser")
    title = soup.find("title").text
    titles.append(title)
    print(f"{i=}, {title=}")
    i = i+1
    time.sleep(np.random.randint(low=1, high= 999)/1000)
```

```python
len(titles)
```

```python
titles = [title.split(" ")[0] for title in titles]
titles = [title.split("")[0] for title in titles]
titles
```

```python
titles = titles[:-1]
titles
```

```python
df = pd.DataFrame(data=titles, columns=["name"])
```
